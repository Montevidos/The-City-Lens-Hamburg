```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# 1. Загружаем CSV
df = pd.read_csv("app_bezirke_EPSG_4326(edited).csv", delimiter=",")

# 2. Удаляем пустые строки geom
df = df[df["geom"].notnull()]

# 3. Функция безопасного парсинга
def safe_load_wkt(val):
    try:
        return wkt.loads(val)
    except:
        return None

# 4. Конвертируем WKT → geometry
df["geometry"] = df["geom"].apply(safe_load_wkt)

# 5. Убираем строки, которые не удалось загрузить
df = df[df["geometry"].notnull()]

# 6. Создаём GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")

# 7. Сохраняем GeoJSON (идеально для Tableau)
gdf.to_file("hamburg_bezirke.geojson", driver="GeoJSON")

print("Готово! → hamburg_bezirke.geojson")
gdf.head()
```
