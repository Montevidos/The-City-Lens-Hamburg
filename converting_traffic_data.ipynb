```python
import pandas as pd
import sys, os
from db_connection import start_engine

engine = start_engine()
```

```python
def scraping_pages(year):

    area = pd.read_excel("Traffic data/T12_Fl√§chennutzung.xlsx", sheet_name=str(year), header=3)
    area.rename(columns={"Unnamed: 0": "Area", 
                        "Unnamed: 1": "Total_land_area",
                        "Unnamed: 2": "Settlement_and_transport",
                        "Siedlung": "Settlement",
                        "darunter Bergbau-betrieb": "incl_mining_operations",
                        "darunter Tagebau, Grube, Steinbruch": "incl_open_pit_mining_pits_quarries",
                        "Verkehr": "Transport",
                        "Unnamed: 7": "Settlement _and_transport_area"
                        }, inplace = True)
    area = area[(area["Area"] == "Metropolregion Hamburg") | (area["Area"] == "Hamburg")]
    area["Year"] = year
    return area

```

```python
years = range(2023, 2017, -1)
saved_df = []
for year in years:
    df = scraping_pages(year)
    saved_df.append(df)
final_df = pd.concat(saved_df, ignore_index=True)
final_df.to_sql(name = "land_using_2018_2023", con = engine, if_exists= "replace", index = False)
```
