```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sys, os
sys.path.append(os.path.dirname(os.getcwd()))
from db_connection import start_engine

engine = start_engine()

url = "https://de.wikipedia.org/wiki/Liste_der_Bezirke_und_Stadtteile_Hamburgs"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, "html.parser")

# pick the table that contains the Bezirke list
tables = soup.find_all("table")

# usually the first large wikitable
table = None
for t in tables:
    if "wikitable" in t.get("class", []):
        table = t
        break

rows = []
for tr in table.find_all("tr"):
    cells = [c.get_text(strip=True) for c in tr.find_all(["th", "td"])]
    rows.append(cells)

df = pd.DataFrame(rows[1:], columns=rows[0])
print(df)
df.to_sql(name = "population_data", con = engine, # Push to DB
                           if_exists= "replace", index = False)
```
